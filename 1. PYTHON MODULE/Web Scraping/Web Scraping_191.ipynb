## Problem Statement
 - Analysis on Mobile Prices

Processor, RAM, Storage, Camera, Battery, Model, Brand, COlor, Price, Screen Size


```python
page = requests.get("https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=52294e36-22ff-45f3-abdb-30a8d995f78a&sort=popularity")
```

```python
page
```

```python
soup = BeautifulSoup(page.text)
```

```python
p = soup.find_all("div",class_="_30jeq3 _1_WHN1")
p
```

```python
prices = []
for i in p:
    prices.append(i.text)
```

```python
len(prices)
```

```python
soup.find_all("div",class_="_4rR01T")[0].text.split()[0]
```

```python
brands = []
for i in soup.find_all("div",class_="_4rR01T"):
    brands.append(i.text.split()[0])
```

```python
soup.find_all("div",class_="_4rR01T")[0].text.split("(")[0]
```

```python
models=[]
for i in soup.find_all("div",class_="_4rR01T"):
    models.append(i.text.split("(")[0])
```

```python
soup.find_all("div",class_="_4rR01T")[0].text.split(",")[1].split()[0]
```

```python
color=[]
for i in soup.find_all("div",class_="_4rR01T"):
    color.append(i.text.split("(")[1].split(",")[0])
```

```python
ROM=[]
for i in soup.find_all("div",class_="_4rR01T"):
    ROM.append(i.text.split(",")[1].split()[0])
```

```python
ROM
```

```python
data = soup.find_all("li",class_="rgWa7D")
```

```python
data[15].text
```

```python
import re
re.findall(r"(\d+)\sGB\sRAM",data[15].text)[0]
```

```python
RAM = []
for i in data:
    a = i.text
    b = re.findall(r"(\d+)\sGB\sRAM",a)
    if len(b)>0:
        RAM.append(b[0])
```

```python
data[1].text
```

```python
re.findall(r"(\d+.\d+)\sinch",data[1].text)[0]
```

```python
ss=[]
for i in data:
    a = i.text
    b = re.findall(r"(\d+.\d+)\sinch",a)
    if b:
        ss.append(b[0])
```

```python
data[18].text
re.findall(r"(\d+)\smAh",data[18].text)[0]
```

```python
battery=[]
for i in data:
    a = i.text
    b = re.findall(r"(\d+)\smAh",a)
    if b:
        battery.append(b[0])
```

```python
battery
```

```python
import pandas as pd
```

```python
dictionary = {
    "Brand":brands,
    "Model":models,
    "RAM":RAM,
    "ROM":ROM,
    "Screen Size":ss,
    "Battery":battery,
    "Color":color,
    "Price":prices
}
```

```python
dictionary['Brand']
```

```python
df = pd.DataFrame(dictionary)
df
```

```python
df.to_csv("batch_191.csv")
```
