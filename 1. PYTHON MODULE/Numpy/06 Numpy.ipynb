### NumPy Arrays v/s List
**Similarities between Lists and Arrays:**

- Both are used for storing data.
- Both are mutable.
- Both can be indexed and iterated through.
- Both can be sliced.

**Differences:**

````python
# Python Lists are an ordered sequence of heterogeneous data types/elements.

bag = [1,'potato', True, 10.5]
print(bag)
print(bag[1])
print(type(bag))
````

````python
# In numpy, the array must contain data of same type (homogeneous). 

# int<float<str
arr1 = np.array([(1,2,3),(4,5,6)])
print(arr1,arr1.shape,arr1.dtype, sep = '\n', end='\n')

print('-'*30)

arr1 = np.array([(1,2,3),(4,15.5,6)])
print(arr1,arr1.shape,arr1.dtype, sep = '\n', end='\n')

print('-'*30)

arr1 = np.array([(1,2,3),(4,15.5,'6')])
print(arr1,arr1.shape,arr1.dtype, sep = '\n', end='\n')

print('-'*30)

arr1 = np.array([(1,2,3),(4,15.5,True)])
print(arr1,arr1.shape,arr1.dtype, sep = '\n', end='\n')

print('-'*30)

arr1 = np.array([(1,2,True),(4,15.5,'6')])
print(arr1,arr1.shape,arr1.dtype, sep = '\n', end='\n')
````

### ii. Arrays with linspace()
The `linspace()` function generates an array with evenly spaced values between specified start and end values, using a specified number of elements.

***Syntax:***

np.linspace(Start, End, Number of elements)

````python
# four values between 10 and 13 (included)
array1d = np.linspace(10, 13, 4) 
print(array1d)
print("-" * 10)

# six values between 10 and 13 (included)
# take any two adjacent values the difference will be same
array1d = np.linspace(10, 13, 6) 
print(array1d)
print("-" * 10)

array1d = np.linspace(10, 11, 5)
print(array1d)
print("-" * 10)

array1d = np.linspace(10, 11, 3)
print(array1d)
print("-" * 10)

array1d = np.linspace(10, 11, 9)
print(array1d)
print("-" * 10)
````

### vi. Eye Array
The `eye()` function, returns an array where all elements are equal to zero, except for the k-th diagonal, whose values are equal to one.

````python
# Return a 2-D array with ones on the diagonal and zeros elsewhere.
array1 = np.eye(3, dtype=int)
print(array1)

print("-" * 10)

array2 = np.eye(5, dtype=int)
print(array2, end = '\n\n')

print("-" * 10)

array2 = np.eye(7,k=3, dtype=int)
print(array2, end = '\n\n')
print("-" * 10)
array2 = np.eye(7,k=-3, dtype=int)
print(array2, end = '\n\n')
````

### Numpy Random Numbers
- The np.random.rand method, generates an array with random numbers that are uniformly distributed between 0 and 1.

- The np.random.randn method, generates an array with random numbers that are normally distributed with mean = 0 and sd = 1.

- The np.random.randint method, generates an array with random numbers that are uniformly distributed between 0 and given integer.

- The np.random.uniform method, generates random float values.

````python
# seed option in np.random
np.random.seed(123) # to reproduce the same set of random numbers

# np.random.rand
print(np.random.rand(), end = '\n\n')
print(np.random.rand(5), end = '\n\n')
print(np.random.rand(3,2), end = '\n\n')
print(np.round(np.random.rand(3,2), 3), end = '\n\n')

print('--'*20)
# np.random.randn
print(np.random.randn(6), end = '\n\n')
print(np.random.randn(3,2), end = '\n\n')


# calculate the mean and std
np.round(np.mean(np.random.randn(300,20)))
np.round(np.std(np.random.randn(300,20)))

print('--'*20)
# np.random.randint
# Return random integers from `low` (inclusive) to `high` (exclusive).
print(np.random.randint(5), end = '\n\n') # range is (0->4)
print(np.random.randint(10, 101), end = '\n\n')
print(np.random.randint(10, 101, (3,2)), end = '\n\n')

print('--'*20)
# np.random.uniform
print(np.random.uniform(5), end = '\n\n') # one randome float value between 0.0 to 5.0
print(np.random.uniform(5,7, (2,5)), end = '\n\n') 
````

### Updating an array with `where()`

````python
arr1 = np.arange(50,80).reshape(5,6)
arr2 = np.arange(60,90).reshape(5,6)
````

````python
print(arr1,arr2, sep = '\n\n')
````

````python
# how many even numbers are present in array "a"
count = 0
for x,y in enumerate(arr1):
    print(f'Iteration {x+1}')
    print(y)
    for w,z in enumerate(y):
        print(f'Iteration {x+1}.{w+1}')
        if (z%2 == 0):
            print(z)
            count = count+1
    print('-'*10)
    
print(count)
````

````python
# solvling the same problem using np.where()

print(np.where(arr1%2==0,'Hi','Bye'))

print(np.where(arr1%2==0,True,False))

print(np.where(arr1%2==0,1,0))

print(np.where(arr1%2==0,1,arr1))
````

````python
np.sum(np.where(arr1%2==0,1,0))
````
