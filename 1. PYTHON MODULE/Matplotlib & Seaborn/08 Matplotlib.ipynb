### 1. line Chart

- A line chart is used to show the change in information over time.
- The horizontal axis is usually a time scale; for example, minutes, hours, days, months, or years.
- For example, you could create a line chart that shows the daily earnings of a store for five days. The horizontal axis would include the days of the week, while the vertical axis would have the daily earnings.

### Create a Line chart to show number of  policies issued by month - Year

```python
 # add policy_bind_MonthYear column

# syntax: date_column.dt.year/date_column.dt.month/date_column.dt.day

# claims_df['policy_bind_date'].dt.day
# claims_df['policy_bind_date'].dt.month
# claims_df['policy_bind_date'].dt.year

# You can format your date as you wish
# df['Month_Year'] = df['Date'].dt.strftime('%b-%Y')

# New column to be added
new_col1 = claims_df['policy_bind_date'].dt.strftime('%b-%Y')
new_col2 = claims_df['policy_bind_date'].dt.year

 
# Inserting the column at the
# beginning in the DataFrame
claims_df.insert(loc = 4,
          column = 'policy_bind_MonthYear',
          value = new_col1)
claims_df.insert(loc = 5,
          column = 'policy_bind_Year',
          value = new_col2)
```

```python
claims_df.head()
```

```python
claims_df.policy_bind_Year.value_counts().sort_index()

claims_df.policy_bind_Year.value_counts().sort_index().index
claims_df.policy_bind_Year.value_counts().sort_index().values
```

```python
# create the line chart

yr = claims_df.policy_bind_Year.value_counts().sort_index().index
val = claims_df.policy_bind_Year.value_counts().sort_index().values

plt.plot(yr, val)

plt.show();
```

### Observation: Since 2015 has lmited data, from 1992-2014 the company is issuing 30-50 policies every year.

```python
plt.figure(figsize=(15,3), dpi = 300)

plt.plot(yr, val)

plt.title('Line chart to show number of policies issued by Year', fontsize = 15, color = 'b')
plt.xlabel('Year',fontsize = 15, color = 'black')
plt.ylabel('No. of Policies',fontsize = 15, color = 'black')

plt.xticks(list(yr)) # manage x-axis values
plt.yticks(range(0,60,5)) # manage y-axis values


plt.show();
```

## 1. Line chart

- A line chart is used to show the change in information over time.
- The horizontal axis is usually a time scale; for example, minutes, hours, days, months, or years.
- For example, you could create a line chart that shows the daily earnings of a store for five days. The horizontal axis would include the days of the week, while the vertical axis would have the daily earnings.

```python
# line chart showing stock prices

dates = pd.date_range(start = '2022-01-01', periods=30)
print(dates)

np.random.seed(123)
stockprice = np.round(np.random.uniform(1477.20,1602.95, 30),2)
print(stockprice)
```

```python
plt.figure(figsize=(15,3), dpi = 300)

plt.plot(dates, stockprice)


plt.title('Line chart to show Stock price for 30 days', fontsize = 15, color = 'b')
plt.xlabel('Date',fontsize = 15, color = 'black')
plt.ylabel('Stock price',fontsize = 15, color = 'black')

plt.xticks(list(dates), rotation = 45) # manage x-axis values
plt.yticks(range(1400,1600,25)) # manage y-axis values


plt.show();
```

## 1.1 Stacked Line chart

```python
# line chart showing stock prices

dates = pd.date_range(start = '2021-01-01', periods=12, freq = 'M')
print(dates)

np.random.seed(123)
grocery_sales = np.round(np.random.uniform(150,322, 12),2)
packagedfoods_slaes = np.round(np.random.uniform(75,250, 12),2)
fruits_veg_sales = np.round(np.random.uniform(110,275, 12),2)
```

```python
plt.figure(figsize=(15,7), dpi = 300)

plt.plot(dates, grocery_sales, label = 'grocery_sales',color = '#464196', ls = '--', lw = 5, marker = '.', markersize = '30')
plt.plot(dates, packagedfoods_slaes, label = 'packagedfoods_slaes', color = 'tomato', ls = '-', lw = 3, marker = 'o', markersize = '10')
plt.plot(dates, fruits_veg_sales, label = 'fruits_veg_sales', color = 'teal', ls = ':', lw = 2, marker = 'H', markersize = '10')

plt.title('Line chart to show category wise sales for 12 months', fontsize = 15, color = 'b')
plt.xlabel('Month',fontsize = 15, color = 'black')
plt.ylabel('Total sales',fontsize = 15, color = 'black')

plt.xticks(list(dates)) # manage x-axis values

plt.legend(loc = 'best')

plt.show();
```

## 2. Area chart

```python
dates = pd.date_range(start = '2022-01-01', periods=30)
print(dates)

np.random.seed(123)
stockprice = np.round(np.random.uniform(850.20,1602.95, 30),2)
print(stockprice)
```

```python
plt.figure(figsize=(15,5), dpi = 300)

plt.stackplot(dates, stockprice)


plt.title('Area chart to show Stock price for 30 days', fontsize = 15, color = 'b')
plt.xlabel('Date',fontsize = 15, color = 'black')
plt.ylabel('Stock price',fontsize = 15, color = 'black')

plt.xticks(list(dates), rotation = 45) # manage x-axis values



plt.show();
```

## 2.1 Stacked Area chart

```python
dates = pd.date_range(start = '2021-01-01', periods=12, freq = 'M')
print(dates)

np.random.seed(123)
grocery_sales = np.round(np.random.uniform(150,322, 12),2)
packagedfoods_slaes = np.round(np.random.uniform(75,250, 12),2)
fruits_veg_sales = np.round(np.random.uniform(110,275, 12),2)

print(grocery_sales,packagedfoods_slaes,fruits_veg_sales)

sales_2darr = np.array([grocery_sales,packagedfoods_slaes,fruits_veg_sales]) # 2-D array
```

```python
plt.figure(figsize=(15,7), dpi = 300)

# we need a 2-D array of values for stacked area chart

plt.stackplot(dates, sales_2darr)

plt.title('Area chart to show category wise sales for 12 months', fontsize = 15, color = 'b')
plt.xlabel('Month',fontsize = 15, color = 'black')
plt.ylabel('Total sales',fontsize = 15, color = 'black')

plt.xticks(list(dates)) # manage x-axis values

plt.legend(['Grocery', 'Packaged food', 'Fruits & Veg'], loc = 'best')

plt.show();
```

## 3. Pie chart

Pie charts can be used to show percentages of a whole.

```python
# pie chart to show market share 
mfg = ['Maruti', 'Honda','Mahindra', 'Hyundai', 'Tata', 'Kia']
mkt_share = [43.72,3.54,6.77,14.93,13.83,6.55]
```

```python
plt.figure(figsize=(7,7), dpi = 100)
plt.pie(mkt_share, labels=mfg, autopct='%.2f%%', radius=0.8)
plt.title('Pie chart to show car market share')
plt.legend(loc = 'upper right')
plt.show();
```

## 4. Scatter plot

Scatter plots’ primary uses are to observe and show relationships between two numeric variables.

```python
car_df = pd.read_csv(r'E:\Excel_files/car_data.csv')
```

```python
plt.figure(figsize=(12,5), dpi = 100)
plt.scatter(car_df.engine_size, car_df.city_mpg)
plt.title('Scatter plot - Engine size vs City Mpg')

plt.xlabel('Engine size',fontsize = 10, color = 'black')
plt.ylabel('City Mpg',fontsize = 10, color = 'black')
plt.show();
```

## 4.1 Scatter plot by categories

```python
plt.figure(figsize=(12,5), dpi = 100)
plt.scatter(car_df.engine_size, car_df.city_mpg, label = 'City Mpg')
plt.scatter(car_df.engine_size, car_df.highway_mpg, label = 'Highway Mpg')

plt.title('Scatter plot - Engine size vs (City Mpg & Highway Mpg)')

plt.xlabel('Engine size',fontsize = 10, color = 'black')
plt.ylabel('Mpg',fontsize = 10, color = 'black')
plt.legend()
plt.show();
```

## 5. Bar chart

A bar chart is used when you want to show a distribution of data points or perform a comparison of metric values across different subgroups of your data.

```python
mfg = car_df.make.value_counts().index
count = car_df.make.value_counts().values
```

```python
Max_val = car_df.make.value_counts().index[0]
Min_val = car_df.make.value_counts().index[-1]
```

```python
plt.figure(figsize=(12,5), dpi = 100)
plt.bar(mfg,count)

plt.title('Bar chart - Num of car for each brand')
plt.xlabel('Brand',fontsize = 10, color = 'black')
plt.ylabel('Count',fontsize = 10, color = 'black')
plt.xticks(rotation = 45)

plt.annotate('Most freq brand is: '+ Max_val, (17,30))
plt.annotate('Least freq brand is: '+ Min_val, (17,28))


plt.legend()
plt.show();
```

## 5.1 Horizontal bar chart

```python
plt.figure(figsize=(12,5), dpi = 100)
plt.barh(mfg,count)

plt.title('Bar chart - Num of car for each brand')
plt.xlabel('Brand',fontsize = 10, color = 'black')
plt.ylabel('Count',fontsize = 10, color = 'black')
plt.xticks(rotation = 45)
plt.legend()
plt.show();
```

## 5.2 Stacked bar chart

```python
segment = ['Passenger vehicles', 'Commercial vehicles', '2-Wheeler']
sales2021 = [1500, 2575, 3550]
sales2020 = [1050, 1900, 1990]
```

```python
df1 = pd.DataFrame(list(zip(segment, sales2020, sales2021)), columns=['Segment', 'Sales2020','Sales2021' ])
```

```python
df1
```

```python
plt.figure(figsize=(12,5), dpi = 100)
plt.bar(df1.Segment,df1.Sales2021, label = '2021')
plt.bar(df1.Segment,df1.Sales2020, label = '2020')


plt.title('Stacked Bar chart')
plt.xlabel('Segment',fontsize = 10, color = 'black')
plt.ylabel('Count',fontsize = 10, color = 'black')

plt.legend()
plt.show();
```

## 5.3 Grouped Bar chart

```python
plt.figure(figsize=(12,5), dpi = 100)


position = np.arange(len(df1.Segment))
w = 0.3

plt.bar(position,df1.Sales2021, label = '2021', width = w)
plt.bar(position+w,df1.Sales2020, label = '2020', width = w)


plt.title('Group Bar chart')
plt.xlabel('Segment',fontsize = 10, color = 'black')
plt.ylabel('Count',fontsize = 10, color = 'black')

plt.xticks(position+0.15, ['PV', 'CV', '2-Wh'])

plt.legend()
plt.show();
```

## 7. Box plot

- Box plots are used to show distributions of numeric data values.
- When you want to compare them between multiple groups.
- Very useful to find outliers in the data.

```python
plt.figure(figsize=(12,5), dpi = 100)
plt.boxplot(car_df.engine_size, vert = 0, patch_artist=True, medianprops=dict(color = 'black'))
plt.show();
```
